```python


import math
import matplotlib.pyplot as plt


a = 1664525
c = 1013904223
m = 2**31
seed = 42

def lcg():
    global seed
    seed = (a * seed + c) % m
    return seed / m

def box_muller_converter():
    a = lcg()
    b = lcg()
    r = math.sqrt(-2 * math.log(a))
    theta = 2 * math.pi * b
    z =  r * math.cos(theta)
    return z

def gaussian(mean, std):
    return mean + std * box_muller_converter()

mean = 50
std  = 10

numbers = []
for i in range(300):
    numbers.append(gaussian(mean, std))


plt.figure()
plt.hist(numbers, bins=30, color='steelblue', edgecolor='white')
plt.title('Gaussian Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')



plt.show()
```
